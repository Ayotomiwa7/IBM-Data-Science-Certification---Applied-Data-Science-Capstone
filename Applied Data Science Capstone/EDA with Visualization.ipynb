```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
df = pd.read_csv(path)
df.head()
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

```python
sns.catplot(y="LaunchSite", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Launch Site",fontsize=20)
plt.show()
```

```python
sns.scatterplot(data=df,x='FlightNumber', y='LaunchSite', hue='Class')
plt.xlabel("Flight Number")
plt.ylabel("Launch Site")
plt.show()
```

```python
sns.scatterplot(data=df,x='FlightNumber', y='LaunchSite')
plt.xlabel("Flight Number")
plt.ylabel("Launch Site")
plt.show()
```

```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df,x="Date",y="Class")
plt.show()
```

```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 'LaunchSite', 'Flights', 'GridFins', 'Reused', 'Legs', 'LandingPad', 'Block', 'ReusedCount', 'Serial']]
features.head()
```

```python
features_one_hot = pd.get_dummies(features)
features_one_hot.head()
```

```python
features_one_hot = features_one_hot.astype("float")
```

```python
features_one_hot.to_csv('dataset_part_3.csv', index=False)
```
